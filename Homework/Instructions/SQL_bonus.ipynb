```python
#creating connection. This is just a test account and has been disabled
engine = create_engine('postgresql://scott:tiger@localhost:5432/HR')
conn=engine.connect()
conn
```

```python
#using sql text format
s=text("select title,avg(salary)::numeric as avgSalary from titles t join salaries s on t.emp_no=s.emp_no group by title")
result=conn.execute(s).fetchall()

#convert to Dataframe
df=pd.DataFrame(result)
df1=df.rename(columns={0:"Title",1:"AvgSalary"})
df1["AvgSalary"]=pd.to_numeric(df1["AvgSalary"])

#plot bar chart
ax=df1.plot.bar(x="Title",y="AvgSalary")

# Set a Title and labels
plt.title("Average Salary per Title")
plt.xlabel("Title")
plt.ylabel("Salary")

#final output
plt.show()


```
